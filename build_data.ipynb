```python
def col_rotation(data: pd.DataFrame,
                 id_vars: list[str],
                 var_name: str,
                 value_name: str,
                 value_type=None,
                 var_typ=None):
    columns = data.columns.tolist()
    num_columns = len(columns)

    year_range = range(1970, 2025)
    new_columns = id_vars + list(year_range)[:num_columns - len(id_vars)]

    if len(new_columns) != num_columns:
        raise ValueError(f"Length mismatch: Expected {num_columns} columns, but got {len(new_columns)} columns.")

    data.columns = new_columns
    data = data.melt(id_vars=id_vars,
                     var_name=var_name,
                     value_name=value_name)
    if value_type is not None:
        data[value_name] = data[value_name].astype(value_type)
    if var_typ is not None:
        data[var_name] = data[var_name].astype(var_typ)
    return data


import pandas as pd

def remove_nan(df: pd.DataFrame, column: str, b1='Country Code', b2='Year', check=False) -> pd.DataFrame:
    # Ensure column exists in the DataFrame
    if column not in df.columns or b1 not in df.columns or b2 not in df.columns:
        raise ValueError("Column not found in DataFrame")

    notna_mask = df[column].notna()
    isna_mask = df[column].isna()

    countries = df[b1].unique()
    for country in countries:
        country_mask = (df[b1] == country)

        # Skip countries if check is enabled and there are non-null values
        check2 = df[country_mask][df[column].notna()]
        if check2.empty or check:
            continue

        country_data = df[country_mask].sort_values(b2)
        notna_country_data = country_data[notna_mask & country_mask]

        for index, row in country_data[isna_mask & country_mask].iterrows():
            year = row[b2]
            if year == 2023:
                prev_year_data = notna_country_data[notna_country_data[b2] == 2022]
                if not prev_year_data.empty:
                    avg = prev_year_data.iloc[0][column]
                else:
                    avg = None
            elif year == 1960:
                next_year_data = notna_country_data[notna_country_data[b2] >= 1960]
                if not next_year_data.empty:
                    avg = next_year_data.iloc[0][column]
                else:
                    avg = None
            else:
                next_year_data = notna_country_data[notna_country_data[b2] > year]
                previous_year_data = notna_country_data[notna_country_data[b2] < year]

                if not next_year_data.empty and not previous_year_data.empty:
                    avg = (next_year_data.iloc[0][column] + previous_year_data.iloc[0][column]) / 2
                elif not previous_year_data.empty:
                    avg = previous_year_data.iloc[0][column]
                elif not next_year_data.empty:
                    avg = next_year_data.iloc[0][column]
                else:
                    avg = notna_country_data[column].mean() if not notna_country_data.empty else None

            df.at[index, column] = avg

    return df

def convert_to_decimal(value):
    if isinstance(value, str):
        # cleaned_str = value.replace(',', '')
        # decimal_number = cleaned_str[0] + '.' + cleaned_str[1:]
        return float(value.replace(',', ''))
    return value
```
